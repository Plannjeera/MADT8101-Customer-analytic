```python
## Import Libraries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
## Import Data
file_path = '/content/E Commerce Dataset.xlsx'
df_name = pd.read_excel(file_path, sheet_name= 'Data Dict',header=1)
df = pd.read_excel(file_path, sheet_name= 'E Comm')
```

```python
# Select the relevant columns for the boxplot
columns = ['Churn', 'Tenure', 'WarehouseToHome', 'OrderAmountHikeFromlastYear', 'OrderCount','DaySinceLastOrder', 'HourSpendOnApp']
subset_df = df[columns]

# Create a 4x2 grid for subplots
fig, axes = plt.subplots(2, 3, figsize=(10, 12))

# Flatten the axes array for easy indexing
axes = axes.flatten()

# Create boxplots for each variable
for i, column in enumerate(columns[1:]):
    sns.boxplot(data=subset_df, x='Churn', y=column, ax=axes[i])

    # Set the title and labels for each subplot
    axes[i].set_title(f'Boxplot for {column}')
    axes[i].set_xlabel('Churn')
    axes[i].set_ylabel(column)

# Adjust the layout to avoid overlapping
plt.tight_layout()

# Show the plot
plt.show()
```

```python
corr_matrix = df.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
from pycaret.classification import *

# Set up the PyCaret environment
#exp_clf = setup(data=df_without_id, target='Churn',remove_multicollinearity = True,feature_selection = True)
clf = setup(df_without_id, target = 'Churn',
            train_size = 0.8,
            numeric_imputation = 'median',
            categorical_imputation = 'mode',
            remove_multicollinearity = True,
            feature_selection = True)
```

```python
# Compare and train multiple models
best_model = compare_models()

# Train the best model
final_model = finalize_model(best_model)
```

```python
plot_model(final_model, plot = 'confusion_matrix') # Left Hand Side
plot_model(final_model, plot = 'feature') # Right Hand Side
```

```python
plot_model(final_model, plot = 'auc') # Left Hand Side
plot_model(final_model, plot = 'pr') # Right Hand Side
```

```python
predictions = predict_model(final_model, data=df)
```
